```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('brands_dataset.xlsx')
df.head()
```

```python
item=df['Select your Five favorite Brands?'].tolist()

item
```

```python
list_item=[]
for i in item:
    list_item.append(i.split(','))
```

```python
# one hot encodeing
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(list_item).transform(list_item)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head()
```

```python
def convert(x):
    if x==True:
        return 1
    else:
        return 0
```

```python
basket_sets=df.applymap(convert)
basket_sets.head()
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.01, use_colnames=True)
```

```python
rules_mlxtend = association_rules(frequent_itemsets, metric="lift", min_threshold=0.5)
rules_mlxtend.head()
```

```python
rules_mlxtend['antecedents'] = rules_mlxtend.apply(lambda row: list(row.antecedents), axis=1)
rules_mlxtend['consequents'] = rules_mlxtend.apply(lambda row: list(row.consequents), axis=1)
```

```python
rules_mlxtend.to_csv('new_recommendation2.csv')
```
